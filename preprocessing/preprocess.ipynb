```python
import pandas as pd
```

```python
# read with first column as index and convert index to int type
df = pd.read_csv('news_items.csv')
```

```python
# select only the columns we need: title, subtitle, content, source, publishing_date, displayed truthfulness rating, readablity, natural language explanation
df = df[['title', 'subtitle', 'content', 'source', 'publishing_date', 'displayed truthfulness rating', 'readability', 'natural language explanation', 'salient features (highlights)']] 
```

```python
# drop rows with missing values in natural language explanation
df = df.dropna(subset=['natural language explanation'])
```

```python
df
```

```python
# covert displayed truthfulness rating to int type from "XX%" string
df['displayed truthfulness rating'] = df['displayed truthfulness rating'].str.replace('%', '').astype(int)
```

```python
# export to json with utf-8 encoding
df.to_json('../src/data/news_items.json', orient='records', force_ascii=False)
```
